##사식연산
더하기 : +

빼기 : -

곱하기 : *

나누기

*   /
*   // : 몫을 반환 (예5/2 = 2.5)
*   % : 나머지를 반환 (예:5/2 = 1)

거듭제곱 : **

```python
# 더하기
print(3+2)    )
```

```python
# 빼기
print(3-1)
```

```python
#  곱하기
print(3*2)
```

```python
# 나누기
print(7/2)
print(7//2)
print(7%2)
```
