```python
class Person:
    def __init__(self):
            self.hello = "안녕하세요"

    def greeting(self):
        print(self.hello)

    호출 = Person()
    호출.greeting()

#self는 인스턴스 자기 자신을 의미
#여기서 __ 언더바 두개 매개변수 self에 들어가는 값은 Persen()이라고 할 수 있다.
```
