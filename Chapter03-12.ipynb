###시권스 사표명
* 문자열 :

    * 시퀀스형에서 사용할 수 있는 메소드 (count(). index(), split() 등...)
    - 슬라이싱([ : ]. [ : : ])
    - 요소들로 이루어져 가능했던 기능들(요소 합치기. 요소 찾기 등등)
* list
    * 시퀀스형에서 사용할 수 있는 메소드 (count(). index(), split() 등...)
    * 슬라이싱([:]) ,[::]
    * 요소들로 이루어져 가능했던 기능들(요소 합치기, 요소 찾기 능등)

    * 2차원 배열( [3,1]. [2:1])  
* tuple  
    * 리스트처럼 요소를 일렬로 저장하지만 만에 저장된 요소를 변경, 추가. 삭제 할 수 없다

    * tuple의 생성방법은 tuple =값. 값 또는 tuple = (값, 값)
    * 그리고 엄청난 비밀을 숨기고 있다
    * 튜플은 소괄호로 생성한다

* dict(딕셔너리)
    * list, tuple, str, range(범위안에 연속되 데이터를 만드는 시퀀스)

    (예시 : 게임케릭터 힘:30, 지능:20, 페력:40, 민첩:90)

```python
# 중복을 제일 싫어하는 set
사과_리스트 = list("apple")
print(사과_리스트)
사과_튜플 = tuple("apple")
print(사과_리스트)
사과_set = set("apple")
print(사과_리스트)
```

```python
# set() 자료형의 데이터 추가, 삭제, 변경
채소

# set() 함수의 데이터 삽입
채소.add('토마토')
채소
채소.update('깻잎')
채소

# set() 함수의 데이터 삭제
채소.remove('잎')  # 삭제할 데이터 없으면 에러
채소

채소.discard('깻') #에러 안남
채소

채소.pop()
채소
```

```python
# 50t = 집합, 합정합. 교집합, 차지합 기타 등등 -| # 대표적으로 합성합, 고집합, 차시합
채소1={"당근","양파","오이","베추"}
채소2={"양파","오이", "대파", "가지"}

# 합장착(union)
채소1| 채소2
set.union(채소1,채소2)

#교집합(intersection)
set.intersection(채소1,채소2)

#차집합(differance)
set.difference(채소1,채소2)
```
