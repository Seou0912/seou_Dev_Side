```python
#5. if~ else 문법과 주의사항(콜론, 돌어쓰기)
# True일때는 서울 안에 작성된 코드가 실행됩니다.
# False일때는 if문 안에 작성된 코드가 실행됩니다.
개수= int(input("붕어빵 몇개 사실건가요?"))
if 개수 > 3:
    print(" 개당 1800원 입니다.")
else :
    print(" 개당 2000원 입니다.")
```

```python
#6. 11~ else은 호름(작동원리)
개수= int(input("봉어빵 몇개 사실건가요?))
IT 개수 > 3:
    print("개당 1800원 입니다.")
else:
    print("개당 2000원 입니다.")
```
