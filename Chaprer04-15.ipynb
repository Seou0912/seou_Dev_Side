###for문 목차
1. for문(반복문)이란?  
. 프로그래밍 언어에서 반복되는 작업을 간단하게 처리하기 위해서 사용하는 기능

2. for문 문법과 주의사항(톤, 들어쓰기)  
3. for문만들어 보기  
4. for문 작동원리  
5. for문에서 range수 활용하기  
6. for문에서 시퀀스 개체로 반복하기  
7. 예제로 for문 익혀보기  

```python
#2. for문으로 만들어보기
#for 변수 in range(횟수):
# print()

#3. for분 만들어보기
for i in range(5):
    print("반복이 되네 신기해!")
```

```python
#4 for 문의 흐름 확인해보기

for i in range(5):
    print("반복이 되네 신기해!",i)
```

```python
#5 for 문에서 range() 함수 활용하기

for i in range(4,10):
        print(i)
```

```python
for i in range(4,10,2): #2씩 증가
        print(i)
```

```python
for i in range(10,4,-1):
        print(i)
```

```python
for i in reversed(range(10)):
        print(i)
```

```python
#6 for 문에서 시퀀스 객체로 반복하기

과일=["두리아","망고","바나나","멜론"]

for i in 과일:
    print(i)
```

```python
과일=["두리아","망고","바나나","멜론"]

for i in "닥스훈트":
    print(i,end='')
```

```python
과일=["두리아","망고","바나나","멜론"]

for i in reversed("닥스훈트"):
    print(i,end='')
```
