```python
f = open('test.txt', "w")
f.write('Hi There! Hi Hi')
f.close()
```

```python
f = open('test.txt', "w")
massage = ''

for i in range(10):
        massage += "Hi There \n"
f. write(massage)
f.close()
```

```python
f = open('test.txt','r')
line = f.readline()
print(line)
print(type(line))
f.close()
```

```python
f = open('test.txt','r')
while line:
    line = f.readline()
    print(line)
f.close()
```

```python
f = open('test.txt', 'a')
f.write("힘내자")
f.close()
```

```python
list = [10,20,30]

with open('test.txt','a') as f:
    f.write(str(list))
```

```python
list = [10,20,30]
with open('test.txt', 'w') as f:
    f.write('let test =' + str(list))
```

```python
list = ['10','20','30']
with open('test.txt', 'w') as f:
    f.writelines(list)
```
