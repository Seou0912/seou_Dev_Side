```python
# 슬라이싱
로또 =[3,5,15,33,41,44]
#로또 변수에 있는 리스트의 인덱스 1번부터 2번까지의 값을 출력
로또 =[3,5,15,33,41,44]
print(로또[1:3])
# 로또 변수의 마지막 위치 출력
print(로또[-1])
# 반대로 출력
print(로또[::-1])
# 로또 리스트안에 33이 있는지 확인해본다
print(33 in 로또)
# 로또 리스트르르 로또2 리스트와 합쳐본다
로또2= [8,2,45,32,39]
print(로또 + 로또2)
#range를 이용해서 1부터 10사이에 짝수만 들어있는 짝수(변수명) 리스트를 만들어주세요
짝수 = list(range(2,12,2))
print(짝수)
#짝수 리스트에 들어있는 값을 2배늘리고 다시 짝수 변수에 담아줘
짝수 = 짝수  *2
print(짝수)
#짝수 리스트에 들어있는 요소의 개수를 구해주세요
print(len(짝수))
#짝수 리스트에 3번째 인데스를 출력해주세요
print(짝수[3])
#len()항수를 이용해 인덱스 마지막 값을 출력해주세요
print(짝수[len(짝수)-1])
```

```python
# list 값 넣고, 바꾸고, 지우고
mzfood = ["숙주" ,"분모자","마라","소세지","소고기","옥수수면"]
# 값을 넣는 방법 2가지 : append(): 마지막 요소에 값을 넣어, insert() : 원하는 곳에 값을 넣어요
mzfood. append("고수")
print(mzfood)

mzfood. insert (0, "탕호루" )
print(mzfood)

#값 지우고
del mzfood[-1]
print(mzfood)

#값 바꾸고
mzfood[-2] ="양고기"
print(mzfood)
```

```python
상자 = {"아디다스": ["신발", "2개"], "나이키": ["모자", "5개"], "뉴발란스": ["가방", "1개"]}
입력할_데이터 = "칼하트 / 후드티 / 2벌"

데이터_리스트 = 입력할_데이터.split(" / ")
상품 = 데이터_리스트[0]
수량 = 데이터_리스트[1]
상자["칼하트"] = [상품, 수량]

print(상자)
```
